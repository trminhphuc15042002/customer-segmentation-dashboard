```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Customers.csv')
```

```python
df.info()
```

```python
# Tìm giá trị mode (giá trị phổ biến nhất) của cột Profession
#.mode()[0] để lấy giá trị đầu tiên trong mode.
profession_mode = df['Profession'].mode()[0]
```

```python
#  Điền các giá trị NaN bằng giá trị của mode
df['Profession'].fillna(profession_mode,inplace=True)
```

```python
# Loại bỏ trùng lặp
# drop_duplicates() loại bỏ các hàng trùng lặp.
# subset=['CustomerID'] chỉ xem xét sự trùng lặp trong cột CustomerID.
# keep='first' giữ lại bản ghi đầu tiên và loại bỏ tất cả các bản sao khác.
# inplace=True áp dụng thay đổi trực tiếp lên DataFrame df.
df.drop_duplicates(subset=['CustomerID'],keep='first',inplace=True)
```

```python
# Xác định Q1,Q3 và IQR
Q1 = df['Annual Income ($)'].quantile(0.25)
Q3 = df['Annual Income ($)'].quantile(0.75)


IQR = Q3 - Q1

# Xác định Outlier
outliers = df[(df['Annual Income ($)'] < (Q1 - 1.5 * IQR)) | (df['Annual Income ($)'] > (Q3 + 1.5 * IQR))]
print("Số lượng ngoại vi: ",len(outliers))
```

```python
# Tạo cột Income Group
conditions_income = [
    (df['Annual Income ($)'] < 50000),
    (df['Annual Income ($)'] >= 50000) & (df['Annual Income ($)'] <= 100000),
    (df['Annual Income ($)'] > 100000)
]

#Định nghĩa nhãn
income_labels = ['Low','Medium','High']

#numpy.select. default = 'Unknown' <=> Else 'Unknown'
df['Income Group'] = np.select(conditions_income,income_labels,default='Unknown')
```

```python
# Tạo cột Speding Category
conditions_spending =[
    (df['Spending Score (1-100)'] <= 40),
    (df['Spending Score (1-100)'] > 40) & (df['Spending Score (1-100)'] <= 70),
    (df['Spending Score (1-100)'] > 70)
]

#Định nghĩa nhãn
spending_labels = ['Low','Medium','High']

df['Speding Category'] = np.select(conditions_spending,spending_labels,default='Unknown')
```

```python
df.to_csv('Customers.csv',index=False)
```
